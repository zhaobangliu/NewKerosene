```python
import glob
import os
import pandas as pd
```

```python

folderPath = "C:\\Users\\zhaobang\\OneDrive\\桌面\\天気データ20220725"
all_files = glob.glob(os.path.join(folderPath, "*.csv"))
df = pd.concat((pd.read_csv(f, encoding="shift-jis") for f in all_files), axis=1)
```

```python
df = df.loc[:,~df.columns.duplicated()]
```

```python
# Change it to a specific order, because we rearraged those coulums
# date,atemperature,ltemperature,sunshine,sonw(accumulated),htemperature,snow,windSpeed(m/s),humidity(%)
cols = ['date', 'atemperature', 'ltemperature', 'sunshine', 'sonw(accumulated)', 'htemperature', 'snow', 'windSpeed(m/s)', 'humidity(%)']
```

```python
df_rearranged = df.copy()[cols]
```
